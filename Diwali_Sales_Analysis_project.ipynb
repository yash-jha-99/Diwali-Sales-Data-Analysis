### Exploratory data Analysis

##### Gender

```python
## Ploting a bar chart for count of Genders
ax=sns.countplot(x='Gender',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
## Plotting a bargraph for Gender vs total count
sales_gender=df.groupby(['Gender'],as_index=False) ['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x='Gender',y='Amount', data=sales_gender)
```

#### Age

```python
ax=sns.countplot(x='Age Group', data=df, hue='Gender')
for bars in ax.containers:
    ax.bar_label(bars)
    
```

```python
sales_age=df.groupby(['Age Group'], as_index=False) ['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x='Age Group', y='Amount', data=sales_age)
```

#### State

```python
## Total number of orders from top 10 states
sales_states= df.groupby(['State'], as_index=False) ['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x='State',y='Orders',data=sales_states)
```

```python
## Total amount/Sales from top 10 states
sales_amount=df.groupby(['State'], as_index=False) ['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x='State',y='Amount', data=sales_amount)
```

#### Marital Status

```python
df.columns
```

```python
# Count of the Marital Status
ax=sns.countplot(data=df,x='Marital_Status')
sns.set(rc={'figure.figsize':(10,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
amount_spent=df.groupby(['Marital_Status','Gender'],as_index=False) ['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(7,5)})
sns.barplot(x='Marital_Status',y='Amount',data=amount_spent,hue='Gender')
```

#### Occupation

```python
ax=sns.countplot(x='Occupation',data=df)
sns.set(rc={'figure.figsize':(25,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
amount_domain=df.groupby(['Occupation'],as_index=False) ['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Occupation',y='Amount',data=amount_domain)
```

#### Product Category

```python
ax=sns.countplot(x='Product_Category',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sns.set(rc={'figure.figsize':(35,8)})
category_amount=df.groupby(['Product_Category'],as_index=False) ['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Product_Category',y='Amount',data=category_amount)
```

```python
# Now I want to see the most sold products top 10
sns.set(rc={'figure.figsize':(20,5)})
product_order=df.groupby(['Product_ID'],as_index=False) ['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.barplot(x='Product_ID',y='Orders',data=product_order)
```

```python
# Top 10 most sole products (same thing as above)
fig1,ax1= plt.subplots(figsize=(12,7))
df.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending=False).plot(kind='bar')
```
